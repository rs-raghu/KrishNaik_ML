```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
day_grouped = df.groupby(by='day')['total_bill'].mean()
```

```python
day_grouped.idxmax()
```

```python
df['tip_pct'] = df['tip']/df['total_bill']*100
df['tip_pct'].head()
```

```python
vip_rows = df[
    (df['total_bill'] > 40) & 
    (df['sex'] == 'Female') & 
    (df['smoker'] == 'No')
]

len(vip_rows)
```

```python
sns.scatterplot(data = df,x='total_bill',y='tip',hue='time')
```
